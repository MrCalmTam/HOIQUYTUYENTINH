```python
import pandas as pd 
wl = '/content/50_Startups.csv'
data=pd.read_csv(wl)
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
corelation = data.corr()
print(corelation)
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(corelation,annot=True)
```

```python
data.plot(kind='scatter',x='R&D Spend',y='Profit',figsize=(9,9),color='red')
```

```python
from sklearn import linear_model
model =linear_model.LinearRegression()
X = pd.DataFrame(data['R&D Spend'])
y = data['Profit']
model.fit(X ,y)
print('Gia tri Intercept la:',model.intercept_)
print('Gia tri Coeffient la:',model.coef_)
```

```python
model.score(X,y )
```

```python
data.plot(kind='scatter',x='R&D Spend',y='Profit',figsize=(9,9),color='red')
y_pred=model.predict(X)
plt.plot(data['R&D Spend'],y_pred,color='blue')
```

```python
X = pd.DataFrame([data['R&D Spend'], data['Profit']**2]).T
model.fit(X,y=data['Profit'])
print('Gia tri Intercept la:',model.intercept_)
print('Gia tri Coeffient la:',model.coef_)
print(' do chinh xac cua mo hinh: ', model.score(X,y))
```

```python
import numpy as np
range=np.arange(0,200000,1)
X_range=pd.DataFrame([range,range**2]).T
y_pred=model.predict(X_range)
data.plot(kind='scatter',x='R&D Spend',y='Profit',figsize=(9,9),color='black',xlim=(0,7))
plt.plot(range,y_pred,color='blue')
```

```python
X = pd.DataFrame([data['R&D Spend'], data['Profit']**3]).T
model.fit(X,y=data['Profit'])
print('Gia tri Intercept la:',model.intercept_)
print('Gia tri Coeffient la:',model.coef_)
print(' do chinh xac cua mo hinh: ', model.score(X,y)) 

import numpy as np
range=np.arange(0,200000,10)
X_range=pd.DataFrame([range,range**3]).T
y_pred=model.predict(X_range)
data.plot(kind='scatter',x='R&D Spend',y='Profit',figsize=(9,9),color='black',xlim=(0,7))
plt.plot(range,y_pred,color='blue')
```
